```python
import numpy as np
import pandas as pd
import yfinance as yf
import scipy.optimize as sco
import matplotlib.pyplot as plt
import pyfolio as pf

# Parameters
N_DAYS = 252
RISKY_ASSETS = ['META', 'TSM', 'ASML', 'IBM']
START_DATE = '2019-05-17'
END_DATE = '2024-05-17'

n_assets = len(RISKY_ASSETS)

# Download stock prices
prices_df = yf.download(RISKY_ASSETS, start=START_DATE, end=END_DATE)['Adj Close']

# Calculate daily returns
returns_df = prices_df.pct_change().dropna()

# Calculate annualized average returns and annualized covariance matrix
avg_returns = returns_df.mean() * N_DAYS
cov_mat = returns_df.cov() * N_DAYS

# Portfolio return function
def get_portf_rtn(w, avg_rtns):
    return np.sum(avg_rtns * w)

# Portfolio volatility function
def get_portf_vol(w, avg_rtns, cov_mat):
    return np.sqrt(np.dot(w.T, np.dot(cov_mat, w)))

# Function to calculate the efficient frontier
def get_efficient_frontier(avg_rtns, cov_mat, rtns_range):
    efficient_portfolios = []
    n_assets = len(avg_rtns)
    args = (avg_rtns, cov_mat)
    bounds = tuple((0, 1) for asset in range(n_assets))
    initial_guess = n_assets * [1. / n_assets, ]
    
    for ret in rtns_range:
        constraints = (
            {'type': 'eq', 'fun': lambda x: get_portf_rtn(x, avg_rtns) - ret},
            {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
        )
        efficient_portfolio = sco.minimize(
            get_portf_vol, initial_guess, args=args,
            method='SLSQP', constraints=constraints, bounds=bounds
        )
        efficient_portfolios.append(efficient_portfolio)
    return efficient_portfolios

# Define the range of expected returns
returns_range = np.linspace(0.0, 0.2, 50)

# Calculate the efficient frontier
efficient_portfolios = get_efficient_frontier(avg_returns, cov_mat, returns_range)

# Extract volatilities from the efficient portfolios
volatilities = [p['fun'] for p in efficient_portfolios]

# Simulate random portfolios for comparison
portf_results = []
for _ in range(10000):
    weights = np.random.random(n_assets)
    weights /= np.sum(weights)
    port_return = np.sum(weights * avg_returns)
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_mat, weights)))
    portf_results.append([port_return, port_volatility, port_return / port_volatility])

# Convert the results to a DataFrame
portf_results_df = pd.DataFrame(portf_results, columns=['returns', 'volatility', 'sharpe_ratio'])

# Plot the efficient frontier and the simulated portfolios
fig, ax = plt.subplots()
portf_results_df.plot(kind='scatter', x='volatility', y='returns', c='sharpe_ratio', cmap='RdYlGn', edgecolors='black', ax=ax)
ax.plot(volatilities, returns_range, 'b--', linewidth=3)
ax.set(xlabel='Volatility', ylabel='Expected Returns', title='Efficient Frontier')

# Identify the minimum volatility portfolio
min_vol_ind = np.argmin(volatilities)
min_vol_portf_rtn = returns_range[min_vol_ind]
min_vol_portf_vol = efficient_portfolios[min_vol_ind]['fun']
min_vol_portf = {
    'Return': min_vol_portf_rtn,
    'Volatility': min_vol_portf_vol,
    'Sharpe Ratio': min_vol_portf_rtn / min_vol_portf_vol
}

# Print the performance summary
print('Minimum volatility portfolio ----')
print('Performance')
for index, value in min_vol_portf.items():
    print(f'{index}: {100 * value:.2f}% ', end="", flush=True)
print('\nWeights')
for x, y in zip(RISKY_ASSETS, efficient_portfolios[min_vol_ind]['x']):
    print(f'{x}: {100*y:.2f}% ', end="", flush=True)

plt.show()

# Pyfolio evaluation
# Select the portfolio with minimum volatility for evaluation
min_vol_weights = efficient_portfolios[min_vol_ind]['x']
min_vol_returns = (returns_df * min_vol_weights).sum(axis=1)

# Evaluate the performance using pyfolio
pf.create_full_tear_sheet(min_vol_returns)
```
